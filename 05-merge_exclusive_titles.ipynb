```python
import pandas as pd
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
```

```python
mangalib = pd.read_csv('data/missing/mangalib_exclusive.csv', sep=';')
remanga = pd.read_csv('data/missing/remanga_matched_mangalib.csv', sep=';')
```

```python
print(mangalib.shape)
mangalib.head(2)
```

```python
print(remanga.shape)
remanga.head(2)
```

```python
remanga[~remanga.mangalib_id.isna()].shape
```

```python
matched_titles = remanga[~remanga.mangalib_id.isna()].mangalib_id.tolist()
mangalib = mangalib[~mangalib.id.isin(matched_titles)]
```

```python
mangalib.shape
```

```python
mangalib_formatted = pd.DataFrame(columns=remanga.columns)

mangalib_formatted.loc[:, 'rus_name'] = mangalib.name
mangalib_formatted.loc[:, 'en_name'] = mangalib.eng_name
mangalib_formatted.loc[:, 'orig_name'] = mangalib.original_name
mangalib_formatted.loc[:, 'type'] = mangalib.manga_type
mangalib_formatted.loc[:, 'dir'] = mangalib.link
mangalib_formatted.loc[:, 'mangalib_id'] = mangalib.id

remanga.loc['orig_name'] = None
exclusive = pd.concat((mangalib_formatted, remanga))
exclusive = exclusive.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
exclusive
```

```python
exclusive = exclusive.drop_duplicates(subset=['dir'])
exclusive
```

```python
exclusive.sample(frac=1).to_csv('data/missing/remanga_mangalib_exclusive.csv', sep=';', index=False)
```
